````python
dane.drop(columns=['Apparent Temperature (C)','Wind Bearing (degrees)','Daily Summary'],inplace=True)
````

````python
def wytnij(fdata,a,b,convert=False):
  if convert:
    return int(fdata[a:b])
  else:
    return fdata[a:b]
dane['data']=dane['Formatted Date'].apply(wytnij,args=(0,10))
dane['rok']=dane['data'].apply(wytnij,args=(0,4,True))
dane['miesiac']=dane['data'].apply(wytnij,args=(5,7,True))

def wytnij_date(fdata):
  return fdata[:10]

# inna wersja
dane['new_date']=pd.to_datetime(dane['Formatted Date'])
dane['year']=dane['new_date'].apply(lambda x:x.year)
````

````python
dane.head()
````

````python
Ddane=dane.groupby('data')
dane_dzienne_avg=Ddane.mean()
dane_dzienne_avg.to_excel('dane_dzienne.xlsx')
dane_dzienne_avg
````

````python
Maj_daily=dane[(dane['rok']==2010) & (dane['miesiac']==5) ].groupby('data')
Maj_minimalne=Maj_daily.min()['Temperature (C)']
Maj_maksymalne=Maj_daily.max()['Temperature (C)']
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.plot(Maj_maksymalne,color='red')
#plt.plot(Maj_minimalne,color='blue')
Maj_maksymalne.plot(kind='line',color='red')
Maj_minimalne.plot(kind='line',color='blue')
plt.xticks(rotation=90)
plt.show()
````

````python
import skimage
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt
plt.set_cmap('gray')
````

````python
image = data.camera()
print(image.shape)
````

````python
#FILTR GAUSSA
image1 = skimage.filters.gaussian(image, sigma=2)
image2 = skimage.filters.gaussian(image1, sigma=5)
image3 = image + 6*(image2-image1)
fig, ax = plt.subplots(1,4,figsize=(20,8)) #jeden wiersz, cztery kolumny obrazków
ax[0].imshow(image)
ax[1].imshow(image1)
ax[2].imshow(image2)
ax[3].imshow(image3)
````

````python
#FILTR SOBELA
im = data.text()
im1 = skimage.filters.sobel(im) 
im2= skimage.filters.sobel_v(im) #sobel pionowy
im3 = skimage.filters.sobel_h(im) #sobel poziomy
fig, ax = plt.subplots(1,4,figsize=(20,8)) #jeden wiersz, cztery kolumny obrazków
ax[0].imshow(im)
ax[1].imshow(im1)
ax[2].imshow(im2)
ax[3].imshow(im3)
````

````python
#SEGMENTACJA
from skimage import exposure
image = data.camera()
histogram=exposure.histogram(image)[0]
prog=filters.threshold_otsu(image)
fig, ax = plt.subplots(1,3,figsize=(20,8))
ax[0].imshow(image)
ax[1].plot(histogram)
ax[1].axvline(prog,color='black')
ax[2].plot(prog)
````
