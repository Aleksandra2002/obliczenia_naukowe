```python
def Lagringo (X,Y,x, symbol):
  w=0
  if symbol:
    sym = ''
  for i in range(len(X)):
    if i>0:
      if symbol:
        sym +='+'
    iloczyn = 1
    if symbol:
      sym += str(Y[i])
    for j in range (len(Y)):
      if i != j:
        iloczyn *= (x-X[j])/(X[i]-X[j])
        if symbol:
          sym += "*(x-"+ str(X[j])+ ")/("+ str(X[i])+ "-"+ str(X[j])+ ")"
    w += Y[i]*iloczyn
  if symbol:
    return w, sym
  else:
    return w
symbol = True  # False

w = Lagringo([1,2,3], [0, 1, 4], 0, symbol)
print("W to :" + str(w[0]) + "wz√≥r:" + w[1])

from sympy import symbols
from sympy.plotting import plot
x= symbols('x')
p1 = plot(str(w[1]), show=False)
from sympy import *
p1.show()
pprint (w[1])
x1=simplify(w[1])
pprint(x1)
#d
w1 =x**5+4*x+3
w2=4*x
table=[2,3,4,1]
tabw1=[]
tabw2=[]
for i in table:
  tabw1.append(w1.subs(x,i))
  tabw2.append(w2.subs(x,i))

w_1= Lagringo(table,tabw1,3,symbol)[1]
w_2= Lagringo (table,tabw2,4, symbol)[1]
plot(w_1,w1)
plot(w_2,w2)
```
