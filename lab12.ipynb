## Apply, groupby, agg

```python
def kategoryzuj_populacje(ludnosc):
  if ludnosc<10000:
    return 'male miasto'
  elif ludnosc<100000:
    return 'srednie miasto'
  else:
    return 'duze miasto'
```

```python
dane['wielkosc miasta']=dane.population.apply(kategoryzuj_populacje) #dodaje kolumne w ktorej dziala funkcja na innej kolumnie
```

```python
def polozenie(wiersz):
  wiersz['polozenie']=f'({wiersz.lat},{wiersz.lng})'
  return wiersz
```

```python
dane.apply(polozenie,axis=1)
```

```python
dane.groupby(['country','admin_name']).max().index #sprawdzanie indeksów
```

```python
dane.agg({'population':['min','max']})
```

```python
plik.drop(columns=['Temperature (C)','Daily Summary'],inplace=True)
```

```python
def wytnij(fdata,a,b,convert=False):
  if convert:
    return int(fdata[a:b])
  else:
    return fdata[a:b]
plik['data']=plik['Formatted Date'].apply(wytnij,args=(0,10))
plik['rok']=plik['data'].apply(wytnij,args=(0,4,True))
plik['miesiac']=plik['data'].apply(wytnij,args=(5,7,True))
```

```python
def wytnij_date(fdata):
  return fdata[:10]
Ddane=plik.groupby('data')
```

```python
dane_dzienne_avg=Ddane.mean()
dane_dzienne_avg.to_excel('dane_dzienne.xlsx')
dane_dzienne_avg
```

```python
majowe=plik[(plik['rok']==2010) & (plik['miesiac']==5)].groupby('data')
maj_min=majowe.min(['Temperature (C)'])
maj_max=majowe.max(['Temperature (C)'])
```
